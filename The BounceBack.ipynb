```python
import pandas as pd
import numpy as np
import winsound
```

```python
time = pd.read_csv("26 weeks after 2.csv")
lows = pd.read_csv("low points 2.csv")
```

```python
time = time.sort_index(ascending=False)
time.head(150)
```

```python
lows = lows.sort_index(ascending=False)
lows
```

```python
limit = 0 
ticker = ""
lowest = 0
close = []
date = []
stock = []
open_price = []
volume = []
num_weeks = []
counter = 0

for index, row in lows.iterrows():
    counter = counter + 1
    print(counter)
    droptick = row["Ticker"]
    dropvalue = row["Close"]
    dropdate = row["Date"]
    rebound = (dropvalue*1.25)
    droplist = time[(time["Ticker"] == droptick)]
    limit = 0
    num_week = 0
    print(droplist)
    for index, line in droplist.iterrows():
        if(line["Date"]>dropdate):
            if(limit<26):
                limit = limit + 1
                num_week = num_week + 1
                if(line["Close"]>=rebound):
                    close.append(line["Close"])
                    date.append(line["Date"])
                    stock.append(line["Ticker"])
                    open_price.append(line["Open"])
                    volume.append(line["Volume"])
                    num_weeks.append(num_week)
                    break
                else:
                    continue
            else:
                break
        else:
            continue

winsound.Beep(440,500)
```

```python
df = pd.DataFrame({
    "Ticker":stock,
    "open":open_price,
    "close":close,
    "volume":volume,
    "date":date,
    "num of weeks":num_weeks
})
df
```
