```python
import pandas as pd
import numpy as np
import winsound
pd.set_option('display.max_rows', 4000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
time = pd.read_csv("26 weeks after 2.csv")
lows = pd.read_csv("low points 2.csv")
```

```python
time = time.iloc[::-1]
time
```

```python
lows = lows.sort_index(ascending=False)
lows
```

```python
limit = 0 
ticker = ""
lowest = 0
close = []
date = []
stock = []
open_price = []
volume = []
num_weeks = []
counter = 0

for index, row in lows.iterrows():
    counter = counter + 1
    print(counter)
    droptick = row["Ticker"]
    dropvalue = row["Close"]
    dropdate = row["Date"]
    rebound = (dropvalue*1.15)
    droplist = time[(time["Ticker"] == droptick)]
    limit = 0
    num_week = 0
    for index, line in droplist.iterrows():
        if(line["Date"]>dropdate):
            if(limit<26):
                limit = limit + 1
                num_week = num_week + 1
                if(line["Close"]>=rebound):
                    close.append(line["Close"])
                    date.append(line["Date"])
                    stock.append(line["Ticker"])
                    open_price.append(line["Open"])
                    volume.append(line["Volume"])
                    num_weeks.append(num_week)
                    break
                else:
                    continue
            else:
                break
        else:
            continue

winsound.Beep(440,500)
```

```python
df = pd.DataFrame({
    "Ticker":stock,
    "open":open_price,
    "close":close,
    "volume":volume,
    "date":date,
    "num of weeks":num_weeks
})
df
```

```python
new_row = ""
new_close = 0
for index, row in df.iterrows():
    ticker = row["Ticker"]
    close = row["close"]
    if(ticker == new_row):
        if(close == new_close):
            df.drop(index, inplace=True)
    if(row["num of weeks"]==26):
        df.drop(index, inplace=True)
    new_row = row["Ticker"]
    new_close = row["close"]
```
