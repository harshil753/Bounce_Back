```python
pricedrop["Date"] = pricedrop["Date"].str.replace('-','')
pricedrop["Date"] = pricedrop["Date"].apply(pd.to_numeric, errors='coerce')
```

```python
pricedrop = pricedrop[pricedrop['Date'] > 20079999]
len(pricedrop)
```

```python
index_values=[]
counter = 0
ticker = []

for index, row in pricedrop.iterrows():
    dropdate=row["Date"]
    droptick=row["Ticker"]
    counter = counter+1
    print(counter)
    droplist = timeseries[(timeseries["Ticker"]==droptick)]
    for index, line in droplist.iterrows():
        if (dropdate == line["Date"]):
            ticker.append(line["Ticker"])
            index_values.append(line["Unnamed: 0"])
import winsound
duration = 1000  # millisecond
freq = 440  # Hz
winsound.Beep(freq, duration)
```

```python
index_df = pd.DataFrame({
    "Ticker": ticker,
    "Index": index_values
})
```

```python
index_df
```

```python
len(index_df)
```

```python
cur = 0
prev = 0
for index, row in index_df:
    cur = row["Index"]
    if((cur - prev)<=4):
        index_df.drop(index)
    prev = cur
```

```python
len(index_df)
import winsound
duration = 1000  # millisecond
freq = 440  # Hz
winsound.Beep(freq, duration)
```

```python
op = []
cl = []
da = []
ti = []
vo = []

counter = 0
for item in index_values:
    counter = counter +1
    print(counter)
    prev = item - 26 
#     fut = item + 26
#     for index, row in pricedrop.iterrows():
#         droptick=row["Ticker"]
#         droplist = timeseries[(timeseries["Ticker"]==droptick)]
    for index, row in timeseries.iterrows():
        for x in range(prev, item):
            if(row["Unnamed: 0"] == x):
                op.append(row["Weekly Open"])
                cl.append(row["Weekly Close"])
                da.append(row["Date"])
                ti.append(row["Ticker"])
                vo.append(row["Volume"])
```
