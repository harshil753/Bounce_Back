```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("the final countdown 2.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
X = df[["close","volume","Market Cap"]]
y= df["num of weeks"].reshape(-1,1)

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1235)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create the model using the MLP Classifier
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(hidden_layer_sizes=(30,), max_iter=1000, random_state=1)
```

```python
model.fit(X_train_scaled, y_train_scaled.ravel())
model
```

```python
training_score = model.score(X_train_scaled, y_train_scaled)
testing_score = model.score(X_test_scaled, y_test_scaled)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)

training_score = model.score(X_train,y_train)
testing_score = model.score(X_test, y_test)

print("m: ", model.coef_)
print("b: ", model.intercept_)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
