```python
import pandas as pd
import numpy as np
import requests as req
import sys
key = "UADYC1EWP37WBZQO"
```

```python
# list of nyse companies
nyse = pd.DataFrame(pd.read_csv("nyse.csv"))
nyse = nyse[["Symbol","Name","MarketCap","Sector","industry"]]
nyse.head()
```

```python
for index, row in nyse.iterrows(): 
    if(("^" or ".") in row["Symbol"]):
        nyse.drop(index, inplace = True)
    
```

```python
new_nyse = nyse.sample(n=650)
```

```python
new_nyse
```

```python
market_cap = []
stock = []
counter = 0 
for index, row in new_nyse.iterrows():
    url = "https://api.iextrading.com/1.0/stock/"+row["Symbol"]+"/quote"
    counter = counter +1
    print(counter)
    print(row["Symbol"])
    print(url)
    print("_______________________________________________")
    try:
        info = req.get(url).json()
        market_cap.append(info["marketCap"])
        stock.append(row["Symbol"])
    except:
        pass
```

```python
marketcap_df = pd.DataFrame({
    "Ticker": stock,
    "Market Cap":market_cap
})
marketcap_df=marketcap_df[["Ticker","Market Cap"]]
marketcap_df
```

```python
for index, row in marketcap_df.iterrows():
    if(row["Market Cap"]==0):
        marketcap_df.drop(index, inplace=True)
marketcap_df
```

```python
# amex monthly data api call
date=[]
weekly_open=[]
weekly_close=[]
weekly_volume = []
ticker=[]
counter=0
for index, row in marketcap_df.iterrows():
    url = "https://www.alphavantage.co/query?"
    counter=counter+1
    function = "TIME_SERIES_WEEKLY"
    symbol = row["Ticker"]
    target_url = url+"function="+function+"&symbol="+symbol+"&apikey="+key
    print(symbol)
    print(counter)
    print(target_url)
    print("______________________________________________________________________")
    try:
        info = req.get(target_url).json()
        for item in info["Weekly Time Series"]: 
                ticker.append(symbol)
                date.append(item)
                weekly_open.append(info["Weekly Time Series"][item]["1. open"])
                weekly_close.append(info["Weekly Time Series"][item]["4. close"])
                weekly_volume.append(info["Weekly Time Series"][item]["5. volume"])
    except:
        pass
```

```python
pricing_df = pd.DataFrame({
    "Ticker":ticker,
    "Weekly Open": weekly_open,
    "Weekly Close": weekly_close,
    "Date":date,
    "Volume":weekly_volume
})
pricing_df
```
