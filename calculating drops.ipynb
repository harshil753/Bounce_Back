```python
import pandas as pd
import numpy as np
import requests as req
```

```python
nyse = pd.read_csv("Pricing Time Series.csv")
```

```python
nyse.drop("Unnamed: 0", axis =1, inplace =True)
nyse
```

```python
date =[] 
open_price =[]
close_price = []
ticker = []
volume =[]

for index, row in nyse.iterrows():
    start = row["Weekly Open"]
    end = row["Weekly Close"]
    ten_perc = (start*.1)
    if((start-end)>=ten_perc):
        date.append(row["Date"])
        open_price.append(start)
        close_price.append(end)
        ticker.append(row["Ticker"])
        volume.append(row["Volume"])
```

```python
df = pd.DataFrame({
    "Opening Price":open_price,
    "Closing Price":close_price,
    "Ticker":ticker,
    "Date":date
})
drops = df[["Ticker","Opening Price","Closing Price","Date"]]
drops
```
